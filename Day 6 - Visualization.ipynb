## Bivariate Data

### a.Line Plot

```python
x = [1,2,3,4,5,6,7]
y = [1,4,9,16,25,36,49]
z = [1,8,27,64,125,216,343]
```

```python
plt.title('Line Plot', fontsize=15)
plt.xlabel('X - Values')
plt.ylabel('Y - Values')

plt.plot(x,y, color = 'midnightblue',
         linestyle='dotted',
         marker = '^',
         markersize = 12,
         markerfacecolor = '#F7DC6F'
        )
plt.show()

```

```python
plt.title('Line Plot', fontsize=15)
plt.xlabel('X - Values')
plt.ylabel('Y - Values')

plt.plot(x,y, color = 'midnightblue',
         linestyle='dotted',
         marker = '^',
         markersize = 10,
         markerfacecolor = '#F7DC6F',
         label = 'Ind'
        )
plt.plot(x,z, color = 'green',
         linestyle='solid',
         marker = 'H',
         markersize = 10,
         markerfacecolor = '#F7DC6F',
         label = 'Pak'
        )
plt.legend()
plt.show()
```

## Numeric Vs Categorical

```python
courses = ['Data Science','Data Analytics','Tableau','Power BI','MySql']
students = [70,57,63,81,45]
```

```python
plt.figure(figsize=(8,4))
plt.bar(courses,students,
       color='#82E0AA',
       alpha = 0.5)
plt.xticks(fontsize=10,
           #rotation=90
          )
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.barh(courses,students,
       color='#EDBB99',
       alpha = 0.5)
plt.xticks(fontsize=10,
           #rotation=90
          )
plt.show()
```

### Pie Plot

```python
plt.pie(students, labels=courses,
       autopct = '%.1f%%',
       textprops={'fontsize':10,'color':'black'},
       explode=(0,0,0,0,0.1))
plt.show()
```

## Catplot

#### Bar Plot

```python
df_cars = pd.read_csv('bar_chart_data.csv')
df_cars
```

```python
pal = sns.color_palette('viridis')
sns.catplot(data=df_cars,x='Brand',y='Cars Listings', 
            kind='bar',
            aspect = 2/1,
            palette = 'winter'
           )
plt.show()
```

```python
sns.catplot(data=df_cars,y='Brand',x='Cars Listings', 
            kind='bar',
            #aspect = 2/1,
            palette = 'winter'
           )
plt.show()
```

#### Box Plot

```python
sns.catplot(data=tips, x = 'day', y='total_bill',
            kind = 'box')
plt.show()
```

```python
sns.catplot(data=tips, x = 'day', y='total_bill',
            kind = 'box', 
            hue = 'time')
plt.show()
```

#### strip plot

```python
sns.catplot(data=tips, x = 'day', y='total_bill',
            kind = 'strip', 
            hue = 'sex'
           )
plt.show()
```

#### Swarm Plot

```python
sns.catplot(data=tips, x = 'day', y='total_bill',
            kind = 'swarm', 
            #hue = 'time'
           )
plt.show()
```
